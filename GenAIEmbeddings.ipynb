```python
import nltk
!pip install transformers
from nltk.corpus import wordnet
import random
from transformers import pipeline

# Ensure required resources are downloaded
nltk.download('wordnet')
nltk.download('omw-1.4')

def get_synonyms(word, topn=3):
    """Retrieve synonyms using WordNet."""
    synonyms = set()
    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            synonyms.add(lemma.name().replace("_", " "))
    return list(synonyms)[:topn]  # Return top N synonyms

# Original prompt
original_prompt = "Write a short story about a futuristic city."

# Keywords to enhance
keywords = ["futuristic", "city"]

# Generate similar words using WordNet
similar_word_mapping = {word: get_synonyms(word) for word in keywords}

def enhance_prompt(prompt, word_mapping):
    """Replace words in the prompt with synonyms."""
    words = prompt.split()
    enriched_words = [random.choice(word_mapping[word]) if word in word_mapping and word_mapping[word] else word for word in words]
    return " ".join(enriched_words)

# Create enriched prompt
enriched_prompt = enhance_prompt(original_prompt, similar_word_mapping)

# Load Hugging Face's GPT-2 text generation model
generator = pipeline("text-generation", model="gpt2")

def generate_response(prompt):
    """Generate text using GPT-2."""
    response = generator(prompt, max_length=100, num_return_sequences=1)
    return response[0]['generated_text']

# Get responses for both prompts
original_response = generate_response(original_prompt)
enriched_response = generate_response(enriched_prompt)

# Print comparisons
print("\nOriginal Prompt:", original_prompt)
print("AI Response:\n", original_response)
print("\nEnriched Prompt:", enriched_prompt)
print("AI Response:\n", enriched_response)

```
